```python
import pandas as pd 
import pycountry_convert as pc
import csv
```

```python
df = pd.read_csv('./transformed_data.csv')
df.dropna()
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.info()
```

```python
df = df[(df[['HDI', 'TC', 'TD', 'STI', 'GDPCAP']] != 0).all(axis=1)]
df.describe()
```

```python
df.isnull().sum()
```

```python
df.interpolate(method='linear', inplace=True)
```

```python
country = df.drop_duplicates(subset=['CODE'], keep='last')
country.info()
```

```python
country = country.reset_index()
country.drop('index', inplace=True, axis=1)
country
```

```python
country_code = country['CODE'].unique().tolist()
alpha2_code = []
for i in country_code:
    alpha2_code.append(pc.country_alpha3_to_country_alpha2(i))
continent_code = []
for i in alpha2_code:
    try:
        continent_code.append(pc.country_alpha2_to_continent_code(i))
    except:
        continent_code.append('Unknown')
        
country['CONTINENT'] = continent_code
country
```

```python
# Data yang sudah dibersihkan
cleaned_data = [country.columns.to_list()] + country.values.tolist()

# Nama file .csv yang baru
new_file_path = './cleaned_country.csv'

# Menulis data ke dalam file .csv baru
with open(new_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(cleaned_data)

print(f"File {new_file_path} berhasil dibuat.")
```
